```python
import ast
```

```python
def convert(text):
    list = []
    
    for i in ast.literal_eval(text):
        list.append(i['name'])
        
    return list
```

```python
movies['genres']
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['cast'].values
```

```python
def convertCast(text):
    list = []
    count = 0
    for i in ast.literal_eval(text):
        if count < 3:
            list.append(i['name'])
            count += 1
        else:
            break
    return list
```

```python
print(movies['cast'].apply(convertCast))
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
movies.head(1)
```

```python
movies['crew'].values
```

```python
def fetchDirector(text):
    list = []
    
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            list.append(i['name'])
            break;
            
    return list
```

```python
movies['crew'].apply(fetchDirector)
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

```python
movies.head(1)
```

```python
def removeSpaces(text):
    list = []
    
    for i in text:
        list.append(i.replace(" ",""))
        
    return list
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast'].values
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
movies = movies.drop(columns = ['overview','keywords', 'genres', 'cast', 'crew'])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['tags'].apply(lambda x:[i.lower() for i in x])
```

```python
movies['tags'] = movies['tags'].apply(lambda x: ' '.join(x))
```

```python
movies.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000 , stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
(cv.get_feature_names_out())
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    list = []
    
    for i in text.split():
        list.append(ps.stem(i))
        
    return list
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
list(enumerate(similarity))
```

```python
sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    mov_index = movies[movies['title'] == movie].index[0]
    dist = similarity[mov_index]
    res = sorted(list(enumerate(dist)),reverse = True , key = lambda x:x[1])[1:6]
    
    for i in res:
        print(movies.iloc[i[0]].title)
        
```

```python
movies.iloc[[0]]
```

```python
recommend("Avatar")
```
